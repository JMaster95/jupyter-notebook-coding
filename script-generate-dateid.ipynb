```python
## mysql create table ddl
CREATE TABLE IF NOT EXISTS `dim_date_day` (
  `date_day_id` VARCHAR(255) COMMENT 'date_day_id',
  `year_week_id` VARCHAR(255) COMMENT 'YEAR_WEEK_ID',
  `year_month_id` VARCHAR(255) COMMENT 'YEAR_MONTH_ID',
  `year_quarter_id` VARCHAR(255) COMMENT 'YEAR_QUARTER_ID',
  `year_id` VARCHAR(255) COMMENT 'YEAR_ID',
  `calendar_date` DATETIME COMMENT 'CALENDAR_DATE',
  `weekday` VARCHAR(255) COMMENT 'WEEKDAY',
  `day_of_month` VARCHAR(255) COMMENT 'DAY_OF_MONTH',
  `month_first_date` DATETIME COMMENT 'MONTH_FIRST_DATE',
  `month_last_date` DATETIME COMMENT 'MONTH_LAST_DATE',
  `quarter_first_date` DATETIME COMMENT 'QUARTER_FIRST_DATE',
  `quarter_last_date` DATETIME COMMENT 'QUARTER_LAST_DATE',
  `year_first_date` DATETIME COMMENT 'YEAR_FIRST_DATE',
  `year_last_date` DATETIME COMMENT 'YEAR_LAST_DATE')
 COMMENT 'Date_day : date';
```

```python
from datetime import date, timedelta
import calendar
# 获取当前日期
current_date = date.today()

# 迭代生成最近10年的日期
for i in range(1, 2):
    # 计算当前年份
    year = current_date.year - i

    # 迭代生成每一天的日期
    for month in range(1, 13):
        for day in range(1, calendar.monthrange(year, month)[1] + 1):
            # 格式化日期为yyyymmdd的格式
            formatted_date = f"{year}{month:02d}{day:02d}"
            print(formatted_date)
```

时间维度表设计
时间维度表应该包含以下字段：

date_key：日期键，作为时间维度表的主键。
calendar_date：日历日期。
day_of_week：星期几，比如星期一、星期二等。
day_of_month：当月的第几天。
day_name：日期的名称，比如1号、2号等。
day_type：工作日或非工作日。
week_of_year：当年的第几周。
month：月份。
quarter：季度。
year：年份。
holiday_flag：是否是节假日。
holiday_name：节假日名称。
season：季节，比如春季、夏季等。

```python
import pandas as pd

# 生成日期范围
date_range = pd.date_range(start='2020-01-01', end='2025-12-31', name='calendar_date')

# 创建时间维度表DataFrame
time_dimension_df = pd.DataFrame(date_range, columns=['calendar_date'])

# 添加其他字段
time_dimension_df['day_of_week'] = time_dimension_df['calendar_date'].dt.dayofweek
time_dimension_df['day_of_month'] = time_dimension_df['calendar_date'].dt.day
time_dimension_df['day_name'] = time_dimension_df['calendar_date'].dt.strftime('%d')
time_dimension_df['day_type'] = time_dimension_df['day_of_week'].apply(lambda x: 'Weekend' if x >= 5 else 'Weekday')
time_dimension_df['week_of_year'] = time_dimension_df['calendar_date'].dt.isocalendar().week
time_dimension_df['month'] = time_dimension_df['calendar_date'].dt.month
time_dimension_df['quarter'] = time_dimension_df['calendar_date'].dt.quarter
time_dimension_df['year'] = time_dimension_df['calendar_date'].dt.year
time_dimension_df['holiday_flag'] = 0
time_dimension_df['holiday_name'] = ''
time_dimension_df['season'] = time_dimension_df['calendar_date'].dt.quarter

# 输出时间维度表数据
print(time_dimension_df)
```
