```python
import numpy as np
import pandas as pd

csv_data_path = 'https://raw.githubusercontent.com/turingplanet/pandas-intro/main/public-datasets/small_survey_results.csv'
survey_df = pd.read_csv(csv_data_path)
survey_df
```

```python
survey_df.dtypes
```

```python
survey_df.drop_duplicates(subset=['MainBranch'], keep='first', inplace=True)
```

```python
survey_df.iloc[2:5]
```

```python
for row_index, row_info in survey_df.iterrows(): # 迭代行
    print(row_index) # 打印 index
    print(row_info) # 打印单行的所有列信息
```

```python
# Hobbyist列中不同数据的分布（相当于group by count?） 结果会自动按照数量从大到小排好序
survey_df['Hobbyist'].value_counts()
```
